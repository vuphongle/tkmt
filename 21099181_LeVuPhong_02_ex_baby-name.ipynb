## TÌM HIỂU DỮ LIỆU

### 01. Đọc dữ liệu

```python
import numpy as np
import pandas as pd

# Đọc dữ liệu từ file CSV vào DataFrame
df = pd.read_csv('US_BaBy_Names.csv', header=0)

# In ra vài dòng đầu của DataFrame để kiểm tra
print(df.head())
```

### 02. Hiển thị 10 dòng dữ liệu bất kỳ

```python
df.shape
#(dòng, cột)
```

### 04. Xóa 2 cột 'Unnamed: 0' and 'Id'

```python
df = df.drop(['Unnamed: 0', 'Id'], axis=1)
#axits = 1 xóa cột axits = 0 xóa dòng
```

### 05. Cho biết số lượng bé trai, bé gái?

```python
# Đếm số lượng bé trai và bé gái trong cột 'Gender'
gender_counts = df['Gender'].value_counts()

# normalize=True trong value_counts() - biết tỉ lệ phần trăm của giới tính
# Hiển thị kết quả
print(gender_counts)
```

### 06. Trong bộ dữ liệu có bao nhiêu tên khác nhau?

```python
unique_names_count = df['Name'].nunique()

# Hiển thị kết quả
print(unique_names_count)
```

### 07. Tên nào xuất hiện ít nhất?

```python
name_counts = df['Name'].value_counts()

# Chọn tên có số lần xuất hiện ít nhất
# least_common_name = name_counts.idxmin() -- tìm theo cách này sẽ hiển thị được 1 tên 
# Hiển thị tên xuất hiện ít nhất
# print("Tên xuất hiện ít nhất là:", least_common_name)

# Tìm số lần xuất hiện ít nhất
min_count = name_counts.min()

# Lọc các tên có số lần xuất hiện bằng với số lần xuất hiện ít nhất
least_common_names = name_counts[name_counts == min_count]

# Hiển thị các tên xuất hiện ít nhất
print("Các tên xuất hiện ít nhất là:")
# print(least_common_names)
print(least_common_names.index.tolist())
```

### 08. Top 10 tên được sử dụng nhiều nhất?

```python
name_counts = df['Name'].value_counts()

# Chọn 10 tên được sử dụng nhiều nhất
top_10_names = name_counts.head(10)

# Hiển thị 10 tên được sử dụng nhiều nhất
print("10 tên được sử dụng nhiều nhất là:")
print(top_10_names)
```

## TRỰC QUAN DỮ LIỆU

### 09. Vẽ đồ thị top 10 tên xuất hiện nhiều nhất năm 2014

```python
import matplotlib.pyplot as plt

# Lọc dữ liệu cho năm 2014
df_2014 = df[df['Year'] == 2014]

# Đếm số lần xuất hiện của mỗi tên trong năm 2014
name_counts_2014 = df_2014['Name'].value_counts()

# Chọn top 10 tên được sử dụng nhiều nhất trong năm 2014
top_10_names_2014 = name_counts_2014.head(10)

# Vẽ đồ thị cột cho top 10 tên xuất hiện nhiều nhất trong năm 2014
# đặt kích thước 10x6
plt.figure(figsize=(10, 6))
top_10_names_2014.plot(kind='bar', color='skyblue')
plt.title('Top 10 tên xuất hiện nhiều nhất năm 2014')
plt.xlabel('Tên')
plt.ylabel('Số lần xuất hiện')
# quay các tên trục x một góc 45 độ
plt.xticks(rotation=45)
plt.show()
```

### 10. Vẽ đồ thị thể hiện tỷ lệ nam nữ qua các năm 2004-2014

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lọc dữ liệu cho các năm từ 2004 đến 2014
df_years = df[(df['Year'] >= 2004) & (df['Year'] <= 2014)]

# Tính tỷ lệ nam nữ cho mỗi năm
gender_ratios = df_years.groupby('Year')['Gender'].value_counts(normalize=True).unstack()

# print(gender_ratios)
# Dùng unstack để đảo cột và dòng
# print(gender_ratios)

# Vẽ đồ thị đường cho tỷ lệ nam nữ qua các năm
plt.figure(figsize=(10, 6))
gender_ratios.plot(kind='line', marker='o')
# marker = 'o' để sử dụng hình tròn để đánh dấu các điểm trên đồ thị
plt.title('Tỷ lệ nam nữ qua các năm (2004-2014)')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ')
plt.xticks(df_years['Year'].unique())
plt.legend(['Nữ', 'Nam'], title='Giới tính', loc='best')
plt.grid(True)
plt.show()
```
